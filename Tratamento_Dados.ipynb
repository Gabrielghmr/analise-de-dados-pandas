# Tratamento de Dados

```python
import pandas as pd
```

```python
dataset = r'https://raw.githubusercontent.com/rafaelpuyau/infinity_school/main/ds/datasets/atividade_casa.csv'
loja = pd.read_csv(dataset)
```

```python
loja.head()
```

```python
loja.info()
```

```python
loja['Data da Venda'] = pd.to_datetime(loja['Data da Venda'], yearfirst=True)
```

```python
loja.info()
```

```python
loja.describe()
```

## Localizando Dados

```python
loja.iloc[0:10, :3]
```

```python

loja_cor_index = loja.set_index('Cor')
loja_cor_index.head()
```

```python

loja_cor_index.loc['Amarela', ['Produto', 'Valor', 'Quantidade']]
```

```python
loja.loc[loja['Valor'] < 100]
```

```python

loja.loc[(loja['Data da Venda'].dt.month == 2) & (loja['Data da Venda'].dt.day > 15) & (loja['Data da Venda'].dt.year == 2022)]
```

```python
loja.loc[(loja['Produto'] == 'Calça') | (loja['Produto'] == 'Camiseta')]['Produto'].value_counts
```

```python
loja.loc[loja['Cor'] == 'Azul']
```

```python
loja.loc[(loja['Produto'] == 'Saia')   &  (loja['Cor'] == 'Azul')]["Quantidade"].sum()
```

```python
loja.loc[ (loja['Produto'] == 'Calça')][ 'Quantidade'].sum()
```

```python
loja.loc[(loja['Produto'] == 'Calça') & (loja['Cor'] == 'Azul')]['Quantidade'].sum()
```

```python
loja.loc[(loja['Produto'] == 'Calça') & (loja['Canal'] == 'Loja Virtual')]['Quantidade'].sum()
```

```python

loja.loc[(loja['Produto'] == 'Camiseta') & (loja['Sexo'] == 'Masculino')]['Sexo'].unique()
```
