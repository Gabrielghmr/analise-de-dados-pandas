```python
#  Qual  o estilo que aparece mais?
estilo_mais_aparece = vinhos['style'].mode()[0]
print(f'O estilo que mais aparece  o estilo: {estilo_mais_aparece}')
```

```python
#  Quantas vezes esse estilo aparece?
qnt_aparicoes_vinho = vinhos['style'].value_counts()['White']
print(f'O vinho do estilo "{estilo_mais_aparece}" teve {qnt_aparicoes_vinho} ocorrencias')
```

```python
vinhos['label'].value_counts().head()
```

```python
#  Quais so as 5 etiquetas que mais aparecem?
etiquetas = vinhos['label'].value_counts().head()

print('As etiquetas que mais apareceram, foram:')
for idx, etiqueta in enumerate(etiquetas.index, start=1):
  print(f'{idx} - {etiqueta} ({etiquetas[etiqueta]})')
```

```python
#  Qual  o preo do vinho mais barato?
menor_preco = vinhos['price'].min()
print(f'O preo do vinho mais barato  R${menor_preco}')
```

```python
#  Qual  o preo do vinho mais caro?
maior_preco = vinhos['price'].max()
print(f'O preo do vinho mais caro  R${maior_preco}')
```

```python
# Quais so as 8 safras mais antigas?
safras_mais_antigas = (pd.Series(vinhos['vintage'].unique())
                      .sort_values(ascending=True)
                      .head(8))

print('As safras mais antigas so de:')
for idx, safra in enumerate(safras_mais_antigas, start=1):
  print(f'{idx} - {safra}')
```
